```python
# load the file
filename = r"X:\Data\Sequence_Class\CD-HITTED\4\1650475848.fas.1"
with open(filename,'r') as f:
    data = f.read()
    f.close

data_split = data.split('>')
data_split = data_split[1:]
data_split = [">"+seq for seq in data_split]
identi = []
sequence = []

for seq in data_split:
    something = seq.split('\n',1)
    if len(something) == 1:
        pass
    else:
         identi.append(something[0])
         sequence.append(something[1])

sequence = [seq.replace('\n','') for seq in sequence]


print("got all the values? ", len(identi) == len(sequence))
print("\nlength of the seqences: ", len(sequence))
```
